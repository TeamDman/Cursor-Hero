```python
from pathlib import Path
import matplotlib.pyplot as plt
icons = list(Path("target/app_icons/msedge.exe").glob("*.png"))
print(icons)
```

```python
import numpy as np
from PIL import Image

icons_rgba = []
for icon_path in icons:
    icon = Image.open(icon_path)
    icon_rgba = np.array(icon.convert("RGBA"))
    icons_rgba.append(icon_rgba)
```

```python
# Now that we have the correct list of extracted files, let's load the first image (0.png) and display each RGBA channel.

# Path to the first image (0.png)

for img_0 in icons_rgba:

    # Display each RGBA channel separately
    fig, ax = plt.subplots(1, 5, figsize=(20, 4))

    # Original image
    ax[0].imshow(img_0)
    ax[0].set_title('Original Image')
    ax[0].axis('off')

    # R channel
    ax[1].imshow(img_0[:, :, 0], cmap='Reds')
    ax[1].set_title('R Channel')
    ax[1].axis('off')

    # G channel
    ax[2].imshow(img_0[:, :, 1], cmap='Greens')
    ax[2].set_title('G Channel')
    ax[2].axis('off')

    # B channel
    ax[3].imshow(img_0[:, :, 2], cmap='Blues')
    ax[3].set_title('B Channel')
    ax[3].axis('off')

    # A channel
    ax[4].imshow(img_0[:, :, 3], cmap='gray')
    ax[4].set_title('A Channel')
    ax[4].axis('off')

    # Show the plots
    plt.tight_layout()
    plt.show()
```
