```python
from pathlib import Path
import matplotlib.pyplot as plt
icons = list(Path("target/app_icons/msedge.exe").glob("*.png"))
print(icons)
```

```python
import numpy as np
from PIL import Image

icons_rgba = []
for icon_path in icons:
    icon = Image.open(icon_path)
    icon_rgba = np.array(icon.convert("RGBA"))
    icons_rgba.append(icon_rgba)
```

```python
# Now that we have the correct list of extracted files, let's load the first image (0.png) and display each RGBA channel.

# Path to the first image (0.png)

for img_0 in icons_rgba:

    # Display each RGBA channel separately
    fig, ax = plt.subplots(1, 6, figsize=(20, 4))
    i = 0

    # Original image
    ax[i].imshow(img_0)
    ax[i].set_title('Original Image')
    ax[i].axis('off')
    i += 1

    # BGRA image
    ax[i].imshow(img_0.astype(np.uint8)[:, :, [2, 1, 0, 3]])
    ax[i].set_title('BGRA Image')
    ax[i].axis('off')
    i += 1

    # R channel
    ax[i].imshow(img_0[:, :, 0], cmap='Reds')
    ax[i].set_title('R Channel')
    ax[i].axis('off')
    i += 1

    # G channel
    ax[i].imshow(img_0[:, :, 1], cmap='Greens')
    ax[i].set_title('G Channel')
    ax[i].axis('off')
    i += 1

    # B channel
    ax[i].imshow(img_0[:, :, 2], cmap='Blues')
    ax[i].set_title('B Channel')
    ax[i].axis('off')
    i += 1

    # A channel
    ax[i].imshow(img_0[:, :, 3], cmap='gray')
    ax[i].set_title('A Channel')
    ax[i].axis('off')
    i += 1

    # Show the plots
    plt.tight_layout()
    plt.show()
```
