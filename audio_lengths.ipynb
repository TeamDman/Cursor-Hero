```python
from pathlib import Path
sounds = Path("assets/sounds").glob("*.ogg")
sound_durations = {}
for sound in sounds:
    # load the sound
    import soundfile as sf
    data, samplerate = sf.read(sound)
    # get the length
    length = len(data) / samplerate
    # print the result
    print(f"{sound.name}: {length:.2f} seconds")
    # store the result
    sound_durations[sound.name] = length
print(len(sound_durations))
```

```python
import os

# Define the path to the sounds.rs file
sounds_rs_path = "src/sounds.rs"

# Open the sounds.rs file in append mode
with open(sounds_rs_path, "a") as file:
    file.truncate(0)
    # Write warning header
    file.write("/*\n")
    file.write(" * WARNING: THIS FILE IS AUTOGENERATED\n")
    file.write(" * DO NOT EDIT THIS FILE DIRECTLY\n")
    file.write(" * \n")
    file.write(" */\n")

    # Write the constants for sound durations
    for sound_name, duration in sound_durations.items():
        # Convert the sound name to a valid Rust identifier
        sound_name = os.path.splitext(sound_name)[0].replace(" ", "_")
        file.write(f"const {sound_name.upper()}_DURATION: f32 = {duration:.2f};\n")
```
